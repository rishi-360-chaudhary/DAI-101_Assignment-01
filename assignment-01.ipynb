<h2 style="color: green;">Importing Libraries</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h1 style="color: green;">Printing first 5 Rows</h1>

```python
df=pd.read_csv('imdb-top-1000.csv')
df.head()
```

```python
# printing info
df.info()
```

<h1 style="color: green;">Adding New Column</h1>

```python
# Adding new column of rating_category which tells movies are excellent if imdb>9 and if b/w 8-9 very good and if b/w 7-8 good and bad for less than 7
def rishi(imdb_score):
    if imdb_score>9:
        return 'Excellent'
    elif 8<= imdb_score <=9:
        return 'Very Good'
    elif 7<= imdb_score <=8:
        return 'Good'
    else:
        return 'Bad'

df['rating_category']=df['IMDB_Rating'].apply(rishi)
#printing any 5 rows
df.sample(5)
```

<h2 style="color: red;">(iii)Categorical and Numerical</h2>

<h3 style="color: green;">a)Z-Test</h3>

```python
from scipy.stats import zscore

# Z-Test Example (Hypothetical)
# Assuming 'rating_category' is binary and we want to compare metascores
group1 = df[df['rating_category'] == 'Excellent']['Metascore']
group2 = df[df['rating_category'] == 'Good']['Metascore']

# Calculate the Z-score
z_score = (group1.mean() - group2.mean()) / (df['Metascore'].std() / (len(df) ** 0.5))
print(f"Z-Score: {z_score}")
```

<h3 style="color: green;">b)T-Test</h3>

```python
from scipy.stats import ttest_ind

# T-Test 
t_stat, p_value = ttest_ind(group1, group2)
print(f"T-Statistic: {t_stat}, p-value: {p_value}")
```

<h3 style="color: green;">c)ANOVA</h3>

```python
from scipy.stats import f_oneway

# ANOVA Example 
anova_result = f_oneway(group1, group2)
print(f"F-Statistic: {anova_result.statistic}, p-value: {anova_result.pvalue}")
```
